```python
import os
import tensorflow as tf
from tqdm import tqdm
import pandas as pd
import skimage
from skimage.io import imread,imshow, imsave
from skimage.transform import resize
import matplotlib.pyplot as plt
import sys
from sklearn.model_selection import train_test_split
```

```python
DATA_DIR = '../data/'
REL_IMAGE_PATH = "../data/images-512/"
REL_MASK_PATH = "../data/masks-512/"
image_names = os.listdir(REL_IMAGE_PATH)
mask_names = os.listdir(REL_MASK_PATH)
```

```python
# Data size
len(image_names)
```

```python
train_size = math.floor(len(image_names) * 70 / 100)
train_size
```

```python
test_size = math.ceil(len(image_names) * 30 / 100)
test_size
```

```python
width = 64
height = 64
channel = 3

# #to create X and Y arrays to be later filled with images of both images and masks
x_train = np.zeros( (train_size,height,width,channel),dtype = np.uint8)
y_train = np.zeros( (train_size,height,width,1)      ,dtype = np.bool)

x_test = np.zeros( (test_size, height, width, channel), dtype = np.uint8 )
y_test = np.zeros( (test_size, height, width, 1)      ,dtype = np.bool )
```

```python
img_output_loc= os.path.join(DATA_DIR, "images-"+str(height))
mask_output_loc = os.path.join(DATA_DIR, "masks-"+str(height))

img_output_loc_train = os.path.join(img_output_loc, "train")
img_output_loc_test = os.path.join(img_output_loc, "test")

mask_output_loc_train = os.path.join(mask_output_loc, "train")
mask_output_loc_test = os.path.join(mask_output_loc, "test")
```

```python
if os.path.isdir(img_output_loc):
    print("You might have already created images with this resolution. Path: "+img_output_loc+".")
else:
    os.makedirs(img_output_loc_train)
    os.makedirs(img_output_loc_test)
    os.makedirs(mask_output_loc_train)
    os.makedirs(mask_output_loc_test)
    train_i = test_i = 0
    for n,id_ in tqdm(enumerate(image_names),total=len(image_names)):
        image_path = os.path.join(REL_IMAGE_PATH, id_)
        img = imread(image_path)[:,:,:channel]
        img = resize(img,(height,width),mode ='edge',preserve_range =True)
        
        mask_path = os.path.join(REL_MASK_PATH, id_[:-4]+"_mask.png")
        mask_ = imread(mask_path)
        mask_resized = resize(mask_,(height,width),mode ='edge',preserve_range =True)
        mask_ = np.expand_dims(mask_resized, axis=-1)
        
        if n < train_size:
            x_train[train_i] = img
            y_train[train_i] = mask_
            train_i += 1
            
            # save the image and mask
            new_img_path = os.path.join(img_output_loc_train, id_)
            imsave(new_img_path, img.astype(np.uint8))

            new_mask_path = os.path.join(mask_output_loc_train, id_[:-4]+"_mask.png")
            imsave(new_mask_path, mask_resized.astype(np.uint8))
        else:
            x_test[test_i] = img
            y_test[test_i] = mask_
            test_i += 1
            
            new_img_path = os.path.join(img_output_loc_test, id_)
            imsave(new_img_path, img.astype(np.uint8))

            new_mask_path = os.path.join(mask_output_loc_test, id_[:-4]+"_mask.png")
            imsave(new_mask_path, mask_resized.astype(np.uint8))
```
