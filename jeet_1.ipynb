## Read image (split/smaller) and label names
The assumption here is that images and labels will be stored just outside this repository.

```python
REL_IMAGE_PATH = "../data/images-256/"
REL_MASK_PATH = "../data/masks-256/"
image_names = os.listdir(REL_IMAGE_PATH)
```

```python
def show_random_images(n):
    fig, ax = plt.subplots(n, 2, sharey='row', figsize = (2*n,4*n))
    for i in range(n):
        image_name = image_names[random.randint(0, len(image_names))]
        image_path = os.path.join(REL_IMAGE_PATH, image_name)
        im = plt.imread(image_path)
        ax[i, 0].imshow(im)
        
        mask_path = os.path.join(REL_MASK_PATH, image_name[:-4]+"_mask.png")
        
        msk = plt.imread(mask_path)
        ax[i, 1].imshow(msk)
```

```python
show_random_images(10)
```
